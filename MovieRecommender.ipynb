```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
credits = pd.read_csv("/content/tmdb_5000_credits.csv")
```

Merging both the dataset into one on the basis of Title

```python
movies=movies.merge(credits,on="title")
```

Deleting unnecessary columns from the dataset to simply the process
Usefull columns:
genre
id
keywords
title
overview

crew

```python
movies=movies[['movie_id','title','genres','overview','keywords','cast','crew']]
movies.head()
```

Cleaning the data: duplictes and null value

```python
movies.dropna(inplace=True)
movies.isnull().sum()
movies.duplicated().sum()
```

merging overview keywords to get tags for a movie

```python
movies.iloc[0].genres
```

```python
#string of [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
#to = ['Action','Adventure','Fantasy','Science Fiction']
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
#to use a function the whole column
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#to take the information of the first three main characters of the movie
def convert3(obj):
  l=[]
  count=0
  for i in ast.literal_eval(obj):
    if count !=3:
      l.append(i['name'])
      count+=1
    else:
      break
  return l
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def get_director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['cast']+movies['crew']+movies['overview']+movies['keywords']+movies['genres']
```

```python
movies.head()
```

```python
df = movies[['movie_id','title','tags']]
df['tags']=df['tags'].apply(lambda x: " ".join(x))
df
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df["tags"][0]
```

```python
df["tags"][1]
```

Two Movies are similar if they have a lot of common tags.
Using the Concept of vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = vectorizer.fit_transform(df["tags"]).toarray()
```

```python
vectors
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

Calculating Cosine Distances between each vectors(movies) with each vector(movies)
Cosine distance is inversly proportional to similarites (i.e more distance less similarities)


```python
similarity = cosine_similarity(vectors)
```

To retain the index number(movie index) for all the movies when we sort, we use enumerate

```python
list(enumerate(similarity))
```

```python
def recommend(movie):
  movie_index = df[df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
